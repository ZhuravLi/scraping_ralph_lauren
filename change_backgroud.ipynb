```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def preprocess(cloth, mask):
    blue_bgd_cloth = cloth.copy()
    for i in range(mask.shape[0]):
        for j in range(mask.shape[1]):
            if mask[i][j] == 0:
                blue_bgd_cloth[i][j] = [0, 0, 1]
    return blue_bgd_cloth                
```

```python
mask = np.load('./mask_imgs/mask_0002.npy')
cloth = plt.imread('./cloth_imgs/cloth_0002.png')

blue_bgd_cloth = preprocess(cloth, mask)
```

```python
plt.figure(figsize=(5, 4))
plt.imshow(mask);
```

```python
plt.figure(figsize=(5, 4))
plt.imshow(cloth);
```

```python
plt.figure(figsize=(5, 4))
plt.imshow(blue_bgd_cloth);
```
